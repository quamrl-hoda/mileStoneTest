## 1.1  what is the difference between static and dynamic variables in python?

In Python, the concepts of **static** and **dynamic** variables can be understood in terms of how variables are scoped, managed, and their behavior at runtime. Here's a breakdown of the differences:

### **i. Static Variables**
- **Definition**: Static variables are variables associated with a class rather than any specific instance of the class.
- **Scope**: They are shared among all instances of the class. Any change made to a static variable affects all instances of the class.
- **Declaration**: Static variables are typically declared directly in the class body but outside any instance methods.
- **Usage**: Used for storing values that are common to all instances.

#### Example:
```python
class Example:
    static_var = "I am a static variable"

# Accessing and modifying the static variable
obj1 = Example()
obj2 = Example()

print(obj1.static_var)  # Output: I am a static variable
print(obj2.static_var)  # Output: I am a static variable

# Changing the static variable
Example.static_var = "Static variable updated"
print(obj1.static_var)  # Output: Static variable updated
print(obj2.static_var)  # Output: Static variable updated
```

### **ii. Dynamic Variables**
- **Definition**: Dynamic variables are typically instance-specific and can change per instance. They are created at runtime.
- **Scope**: Dynamic variables belong to individual instances of a class. Changes to a dynamic variable of one instance do not affect others.
- **Declaration**: Declared inside instance methods using the `self` keyword.
- **Usage**: Used for storing data that is unique to each instance of a class.

#### Example:
```python
class Example:
    def __init__(self, value):
        self.dynamic_var = value  # Dynamic variable

# Creating instances with different dynamic variables
obj1 = Example("Dynamic value for obj1")
obj2 = Example("Dynamic value for obj2")

print(obj1.dynamic_var)  # Output: Dynamic value for obj1
print(obj2.dynamic_var)  # Output: Dynamic value for obj2

# Modifying the dynamic variable
obj1.dynamic_var = "Updated for obj1"
print(obj1.dynamic_var)  # Output: Updated for obj1
print(obj2.dynamic_var)  # Output: Dynamic value for obj2 (unchanged)
```

###Explain the purpose of `pop` `popitem`, `clear()` in a dictionary with suitable example.

### i. `pop()`
The `pop()` method removes a specified key from the dictionary and returns its associated value. If the key is not found, it raises a `KeyError` unless a default value is provided.

**Example:**
```python
# Creating a dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3}

# Using pop() to remove the key 'b'
value = my_dict.pop('b')
print("Removed value:", value)  # Output: Removed value: 2
print("Updated dictionary:", my_dict)  # Output: Updated dictionary: {'a': 1, 'c': 3}
```

### ii. `popitem()`
The `popitem()` method removes and returns the last inserted key-value pair as a tuple. This method is useful when you want to remove items in a Last In First Out (LIFO) manner. If the dictionary is empty, it raises a `KeyError`.

**Example:**
```python
# Creating a dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3}

# Using popitem() to remove the last inserted item
key_value = my_dict.popitem()
print("Removed item:", key_value)  # Output: Removed item: ('c', 3)
print("Updated dictionary:", my_dict)  # Output: Updated dictionary: {'a': 1, 'b': 2}
```

### iii. `clear()`
The `clear()` method removes all items from the dictionary, leaving it empty. It does not return any value.

**Example:**
```python
# Creating a dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3}

# Using clear() to remove all items
my_dict.clear()
print("Cleared dictionary:", my_dict)  # Output: Cleared dictionary: {}

   my_list = [1, 2, 3]

   ```python
   my_dict = {'a': 1, 'b': 2}
   my_dict['c'] = 3  # Adding a new key-value pair
   print(my_dict)  # Output: {'a': 1, 'b': 2, 'c': 3}
   ```

 **Sets**: Sets can have elements added or removed.
   ```python
   my_set = {1, 2, 3}

   ```

**Examples of Immutable Data Types:**
 **Tuples**: Once a tuple is created, you cannot change its elements.

   my_tuple = (1, 2, 3)

   ```

### What is docstring in Python?Explain with an example.

A **docstring** in Python is a special type of comment that is used to document a module, class, method, or function. It provides a convenient way of associating documentation with Python code. Docstrings are written as string literals and are typically placed as the first statement in a module, class, or function definition.

### Characteristics of Docstrings:
- **Enclosed in Triple Quotes**: Docstrings can be enclosed in triple single quotes (`'''`) or triple double quotes (`"""`).
- **Accessed via `__doc__` Attribute**: You can access the docstring of an object using its `__doc__` attribute.
- **Used by Documentation Tools**: Docstrings can be automatically extracted by documentation generation tools like Sphinx.

### Purpose of Docstrings:
- To describe what a function, method, or class does.
- To explain the parameters and return values.
- To provide usage examples.

### Example of a Docstring

Here’s an example that demonstrates how to use docstrings in a function:

```python
def add(a, b):
    """
    Add two numbers and return the result.

    Parameters:
    a (int or float): The first number to add.
    b (int or float): The second number to add.

    Returns:
    int or float: The sum of the two numbers.
    
    Example:
    >>> add(2, 3)
    5
    >>> add(2.5, 3.5)
    6.0
    """
    return a + b

# Accessing the docstring
print(add.__doc__)
```

### What is break, continue and pass in Python?
In Python, `break`, `continue`, and `pass` are control flow statements that help manage the execution of loops and conditional statements. Each serves a different purpose:

### `break`
The `break` statement is used to exit a loop prematurely. When encountered, it terminates the nearest enclosing loop (for or while) and transfers control to the next statement following the loop.

**Example of `break`:**
```python
for number in range(10):
    if number == 5:
        break  # Exit the loop when number is 5
    print(number)

# Output:
# 0
# 1
# 2
# 3
# 4
```

###  `continue`
The `continue` statement skips the current iteration of a loop and proceeds to the next iteration. It does not terminate the loop but rather bypasses the remaining code inside the loop for that specific iteration.

**Example of `continue`:**
```python
for number in range(10):
    if number % 2 == 0:
        continue  # Skip even numbers
    print(number)

# Output:
# 1
# 3
# 5
# 7
# 9
```

###  `pass`
The `pass` statement is a null operation; it serves as a placeholder in situations where syntactically some code is required but you do not want to execute any code. It can be useful in defining empty functions, classes, or loops.

**Example of `pass`:**
```python
def function_that_does_nothing():
    pass  # Placeholder for future code

for number in range(5):
    if number < 3:
        pass  # Do nothing for numbers less than 3
    else:
        print(number)

# Output:
# 3
# 4
```


### What is the use of self in Python?
In Python, `self` is a conventional name used to refer to the instance of a class within its methods.  **Instance Reference**: `self` refers to the current instance of the class. It differentiates between instance attributes and local variables.

 **Accessing Attributes and Methods**: Using `self`, you can access attributes and methods of the class within its own methods.

### What are global, protected and private attributes in Python?

In Python, attributes of a class can be categorized into three types based on their accessibility: **global**, **protected**, and **private**. Understanding these attribute types is crucial for encapsulation and data hiding in object-oriented programming.

### Global Attributes
Global attributes are defined outside of any class or function and can be accessed from anywhere in the code, including inside classes and functions. They are not tied to any specific object.

**Example of Global Attributes:**
```python
# Global variable
global_var = "I am a global variable"

class MyClass:
    def print_global(self):
        print(global_var)  # Accessing global variable

# Creating an instance of MyClass
obj = MyClass()
obj.print_global()  # Output: I am a global variable
```

###  Protected Attributes
Protected attributes are defined within a class and are intended to be accessible only within that class and its subclasses. In Python, protected attributes are indicated by a single underscore prefix (`_`). While they can still be accessed from outside the class, it is considered a convention that they should not be.

**Example of Protected Attributes:**
```python
class Base:
    def __init__(self):
        self._protected_var = "I am a protected variable"

class Derived(Base):
    def print_protected(self):
        print(self._protected_var)  # Accessing protected variable

# Creating an instance of Derived
obj = Derived()
obj.print_protected()  # Output: I am a protected variable
```

###Private Attributes
Private attributes are defined within a class and are intended to be accessible only within that class. They are indicated by a double underscore prefix (`__`). This name mangling makes it harder (but not impossible) to access them from outside the class.

**Example of Private Attributes:**
```python
class MyClass:
    def __init__(self):
        self.__private_var = "I am a private variable"

    def get_private(self):
        return self.__private_var  # Accessing private variable within the class

# Creating an instance of MyClass
obj = MyClass()
print(obj.get_private())  # Output: I am a private variable

# Attempting to access private variable directly will raise an AttributeError
# print(obj.__private_var)  # This will raise an AttributeError
```

### What are lists and tuples? What is the key difference between the two?


In Python, **lists** and **tuples** are both used to store collections of items. However, they have some key differences in terms of mutability, syntax, and usage. Here’s a detailed explanation of both data structures along with their differences.

### Lists
- **Definition**: A list is a mutable, ordered collection of items. You can change, add, or remove elements from a list after it has been created.
- **Syntax**: Lists are defined using square brackets `[]`.

**Example of a List**:
```python
# Creating a list
my_list = [1, 2, 3, 4, 5]

# Modifying the list
my_list.append(6)  # Adding an element
print(my_list)     # Output: [1, 2, 3, 4, 5, 6]

# Removing an element
my_list.remove(3)
print(my_list)     # Output: [1, 2, 4, 5, 6]
```

### Tuples
- **Definition**: A tuple is an immutable, ordered collection of items. Once a tuple is created, you cannot change its contents (i.e., you cannot add or remove elements).
- **Syntax**: Tuples are defined using parentheses `()`.

**Example of a Tuple**:
```python
# Creating a tuple
my_tuple = (1, 2, 3, 4, 5)

# Attempting to modify the tuple (will raise an error)
# my_tuple[0] = 10  # This will raise a TypeError

print(my_tuple)     # Output: (1, 2, 3, 4, 5)
```

### Key Differences Between Lists and Tuples

| Feature         | Lists                          | Tuples                       |
|------------------|--------------------------------|------------------------------|
| Mutability       | Mutable (can be changed)      | Immutable (cannot be changed)|
| Syntax           | Defined with square brackets `[]` | Defined with parentheses `()` |
| Performance      | Slower than tuples due to mutability | Faster than lists due to immutability |
| Use Cases        | Used when you need a collection that can change over time | Used for fixed collections of items |
| Methods          | More built-in methods (e.g., `append()`, `remove()`) | Fewer methods available |

### What are Dict and List comprehensions?

In Python, **list comprehensions** and **dictionary comprehensions** are concise ways to create lists and dictionaries, respectively. They allow you to generate these data structures in a single line of code, making your code more readable and efficient.

### List Comprehensions

A list comprehension provides a syntactic construct for creating a new list by applying an expression to each item in an iterable (like a list or range). The basic syntax is:

```python
new_list = [expression for item in iterable if condition]
```

- **expression**: The value to add to the list.
- **item**: The variable that takes the value of each element in the iterable.
- **iterable**: Any iterable object (like a list, tuple, or string).
- **condition**: An optional filter that determines whether the expression should be included.

#### Example of List Comprehension

```python
# Creating a list of squares of even numbers from 0 to 9
squares = [x**2 for x in range(10) if x % 2 == 0]
print(squares)  # Output: [0, 4, 16, 36, 64]
```

### Dictionary Comprehensions

A dictionary comprehension is similar to a list comprehension but creates a dictionary instead. The syntax is:

```python
new_dict = {key_expression: value_expression for item in iterable if condition}
```

- **key_expression**: The expression that defines the key.
- **value_expression**: The expression that defines the value.
- **item**, **iterable**, and **condition** work similarly as in list comprehensions.

#### Example of Dictionary Comprehension

```python
# Creating a dictionary where keys are numbers and values are their squares
squares_dict = {x: x**2 for x in range(5)}
print(squares_dict)  # Output: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
```


### What are decorators in Python? Explain it with an example.Write down its use cases.

In Python, **decorators** are a powerful and expressive tool that allows you to modify or enhance the behavior of functions or methods without changing their actual code. Decorators are often used for logging, enforcing access control, instrumentation, and caching, among other use cases.

### What is a Decorator?

A decorator is a function that takes another function as an argument and extends or alters its behavior. It is typically defined using the `@decorator_name` syntax above the function definition.

### Example of a Decorator

Here’s a simple example to illustrate how decorators work:

```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()  # Call the original function
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

# Calling the decorated function
say_hello()
```

### Use Cases for Decorators

Decorators can be used in various scenarios, including:

1. **Logging**: Automatically log entry and exit points of functions.
2. **Authentication**: Check whether a user has permission to execute a particular function.
3. **Caching**: Store results of expensive function calls and return cached results when the same inputs occur again.
4. **Timing Functions**: Measure how long a function takes to execute.
5. **Input Validation**: Validate inputs before executing a function.


### How is memory managed in Python?

Memory management in Python is a crucial aspect of the language that ensures efficient allocation and deallocation of memory for objects and data structures. Here’s an overview of how memory is managed in Python:

### Key Components of Memory Management in Python

 **Automatic Memory Management**:
   - Python uses automatic memory management, which means that the programmer does not need to manually allocate and free memory. This is handled by the Python interpreter.

 **Garbage Collection**:
   - Python employs a garbage collection mechanism to reclaim memory occupied by objects that are no longer in use. The main technique used is reference counting, where each object maintains a count of references pointing to it.
   - When the reference count drops to zero (i.e., no references to the object exist), the memory occupied by that object can be freed.
   - Additionally, Python has a cyclic garbage collector that can detect and collect objects involved in reference cycles (where two or more objects reference each other).

 **Memory Pools**:
   - Python uses a private heap space for storing all its objects and data structures. The management of this heap is done internally by the Python memory manager.
   - The memory manager allocates memory blocks from the heap for different data types and keeps track of them.

 **Object-Specific Allocators**:
   - For certain built-in types, such as integers and lists, Python uses specialized allocators to optimize memory usage and performance.
   - For instance, small integers are cached to avoid creating multiple instances of the same integer.
 **Memory Fragmentation**:
   - Over time, as objects are created and destroyed, fragmentation can occur in the heap. Python's garbage collector helps mitigate this issue by cleaning up unused objects.

### Example of Memory Management

Here’s a simple example to illustrate how memory management works:

```python
# Creating a list
my_list = [1, 2, 3]

# Creating another reference to the same list
another_reference = my_list

# Deleting one reference
del another_reference

# At this point, my_list still exists because it has one reference
print(my_list)  # Output: [1, 2, 3]

# Deleting the remaining reference
del my_list

# Now both references are gone, and the list can be garbage collected
```

### What are iterators, iterable & generators in Python?


In Python, **iterators**, **iterables**, and **generators** are important concepts related to the handling of sequences and collections of data. Here’s an overview of each:

###  Iterable

An **iterable** is any Python object that can return its elements one at a time. This includes built-in data structures like lists, tuples, sets, dictionaries, and strings. An iterable implements the `__iter__()` method or the `__getitem__()` method.

#### Example of an Iterable:
```python
# A list is an iterable
my_list = [1, 2, 3, 4]

# You can iterate over it using a for loop
for item in my_list:
    print(item)
```

### Iterator

An **iterator** is an object that represents a stream of data; it is created from an iterable. Iterators implement two methods: `__iter__()` (which returns the iterator object itself) and `__next__()` (which returns the next value from the iterator). When there are no more items to return, `__next__()` raises a `StopIteration` exception.

#### Example of an Iterator:
```python
# Creating an iterator from a list
my_list = [1, 2, 3, 4]
my_iterator = iter(my_list)  # Get an iterator from the list

# Using the iterator to get elements one by one
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2

# Continue until StopIteration is raised
try:
    while True:
        print(next(my_iterator))
except StopIteration:
    pass
```

###  Generator

A **generator** is a special type of iterator that is defined using a function with the `yield` statement. Generators allow you to create iterators in a more concise way without needing to implement the iterator protocol (i.e., `__iter__()` and `__next__()`). When a generator function is called, it returns a generator object but does not start execution immediately. Each call to `next()` on the generator resumes execution until it hits `yield`, which provides a value and pauses the function's state.

#### Example of a Generator:
```python
def my_generator():
    for i in range(5):
        yield i * 2  # Yielding values instead of returning

# Creating a generator object
gen = my_generator()

# Using the generator to get values
for value in gen:
    print(value)

# Output:
# 0
# 2
# 4
# 6
# 8
```


### What is the difference between xrange and range in Python?

In Python, `range()` and `xrange()` are both functions used to generate a sequence of numbers, but they have some key differences. However, it's important to note that `xrange()` was available in Python 2.x and is not present in Python 3.x. In Python 3.x, `range()` behaves like `xrange()` did in Python 2.x.

### Differences Between `range()` and `xrange()`

 **Return Type**:
   - **`range()`**: Returns a list of numbers. In Python 2.x, it creates a list containing all the numbers in the specified range.
   - **`xrange()`**: Returns an xrange object, which generates the numbers on demand (lazy evaluation). It does not create a list in memory.

 **Memory Efficiency**:
   - **`range()`**: Since it generates a list, it consumes more memory, especially for large ranges.
   - **`xrange()`**: More memory efficient as it generates numbers one at a time and does not store the entire list in memory.

 **Usage in Loops**:
   - Both can be used in loops, but using `xrange()` is generally preferred in Python 2.x for large ranges due to its efficiency.

 **Python Version Compatibility**:
   - **`range()`**: Available in both Python 2.x and 3.x.
   - **`xrange()`**: Only available in Python 2.x. In Python 3.x, `range()` behaves like `xrange()`, providing the same memory efficiency.

 **Functionality**:
   - Both functions allow you to specify start, stop, and step values, but because `xrange()` is an iterator, you cannot use it where a list is required (e.g., indexing).

### Example Usage

#### In Python 2.x
```python
# Using range()
for i in range(5):  
    print(i)  # Output: 0, 1, 2, 3, 4

# Using xrange()
for i in xrange(5):  
    print(i)  # Output: 0, 1, 2, 3, 4
```

#### In Python 3.x
```python
# Using range() (which behaves like xrange() from Python 2)
for i in range(5):  
    print(i)  # Output: 0, 1, 2, 3, 4

### 1.8. Write a program to find the factors of a whole number using a while loop.


````python
def find_factors(number):
    factors = []
    i = 1  # Start from 1, as factors are positive integers

    while i <= number:  # Loop until i is greater than the number
        if number % i == 0:  # Check if i is a factor
            factors.append(i)  # Add to the list of factors
        i += 1  # Increment i to check the next number

    return factors

# Main program
if __name__ == "__main__":
    # Prompt user for input
    try:
        num = int(input("Enter a whole number to find its factors: "))
        if num < 0:
            print("Please enter a non-negative whole number.")
        else:
            result = find_factors(num)
            print(f"The factors of {num} are: {result}")
    except ValueError:
        print("Invalid input! Please enter a valid whole number.")
````

### 1.10. Write a program to find the sum of all the positive numbers entered by the user. As soon as the user enters a negative number, stop taking in any further input from the user and display the sum.


````python
def sum_of_positive_numbers():
    total_sum = 0  # Initialize the sum to zero

    while True:
        try:
            # Prompt user for input
            number = float(input("Enter a positive number (or a negative number to stop): "))

            if number < 0:  # Check if the number is negative
                break  # Exit the loop if a negative number is entered

            total_sum += number  # Add the positive number to the total sum

        except ValueError:
            print("Invalid input! Please enter a valid number.")

    return total_sum

# Main program
if __name__ == "__main__":
    result = sum_of_positive_numbers()
    print(f"The sum of all positive numbers entered is: {result}")
````

### 1.11. Write a program to find prime numbers between 2 to 100 using nested for loops.

````python
def find_prime_numbers(start, end):
    prime_numbers = []

    for num in range(start, end + 1):  # Iterate through each number in the range
        if num > 1:  # Check only numbers greater than 1
            is_prime = True  # Assume the number is prime

            for i in range(2, int(num**0.5) + 1):  # Check for factors from 2 to sqrt(num)
                if num % i == 0:  # If num is divisible by i, it's not prime
                    is_prime = False
                    break  # No need to check further

            if is_prime:  # If the number is still considered prime
                prime_numbers.append(num)  # Add it to the list of prime numbers

    return prime_numbers

# Main program
if __name__ == "__main__":
    primes = find_prime_numbers(2, 100)
    print("Prime numbers between 2 and 100 are:", primes)
````

- Explain which celestial body (Earth or Moon) is more attracted to the other based on the comparison.

````python
# Constants
G = 6.674e-11  # Gravitational constant in N(m/kg)^2
mass_earth = 5.972e24  # Mass of Earth in kg
mass_moon = 7.34767309e22  # Mass of Moon in kg
mass_sun = 1.989e30    # Mass of Sun in kg
distance_earth_sun = 1.496e11  # Distance between Earth and Sun in meters
distance_moon_earth = 3.844e8    # Distance between Moon and Earth in meters

# Calculate gravitational force between Earth and Sun
F_earth_sun = (G * mass_earth * mass_sun) / (distance_earth_sun ** 2)

# Calculate gravitational force between Moon and Earth
F_moon_earth = (G * mass_moon * mass_earth) / (distance_moon_earth ** 2)

# Output results
print(f"Gravitational Force between Earth and Sun: {F_earth_sun:.2e} N")
print(f"Gravitational Force between Moon and Earth: {F_moon_earth:.2e} N")

# Compare forces
if F_earth_sun > F_moon_earth:
    print("The gravitational force between Earth and Sun is stronger.")
else:
    print("The gravitational force between Moon and Earth is stronger.")
````

### 4.Create a simple banking system using object-oriented concepts in Python. Design classes representing different types of bank accounts such as savings and checking. Implement methods for deposit, withdraw, and balance inquiry. Utilize inheritance to manage different account types efficiently.

Tasks:

- Define base class(es) for bank accounts with common attributes and methods.

- Implement subclasses for specific account types (e.g, SavingsAccount, CheckingAccount).

- Provide methods for deposit, withdraw, and balance inquiry in each subclass.

- Test the banking system by creating instances of different account types and performing transactions.

````python
class BankAccount:
    def __init__(self, account_holder, initial_balance=0):
        self.account_holder = account_holder
        self.balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited: ${amount:.2f}. New balance: ${self.balance:.2f}.")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                print(f"Withdrew: ${amount:.2f}. New balance: ${self.balance:.2f}.")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    def get_balance(self):
        return self.balance


class SavingsAccount(BankAccount):
    def __init__(self, account_holder, initial_balance=0):
        super().__init__(account_holder, initial_balance)

    def withdraw(self, amount):
        # Savings accounts may have withdrawal limits (optional feature)
        if amount > 500:
            print("Withdrawal limit exceeded for Savings Account. Maximum allowed is $500.")
        else:
            super().withdraw(amount)


class CheckingAccount(BankAccount):
    def __init__(self, account_holder, initial_balance=0):
        super().__init__(account_holder, initial_balance)

    def withdraw(self, amount):
        # Checking accounts can withdraw any amount as long as there are sufficient funds
        super().withdraw(amount)


# Main program to test the banking system
if __name__ == "__main__":
    # Creating instances of different account types
    savings_account = SavingsAccount("Alice", 1000)
    checking_account = CheckingAccount("Bob", 500)

    # Testing deposit and withdrawal on Savings Account
    print("\n--- Savings Account Transactions ---")
    savings_account.deposit(200)              # Deposit into savings
    savings_account.withdraw(150)              # Withdraw from savings
    savings_account.withdraw(600)              # Attempt to withdraw more than limit
    print(f"Savings Account Balance: ${savings_account.get_balance():.2f}")

    # Testing deposit and withdrawal on Checking Account
    print("\n--- Checking Account Transactions ---")
    checking_account.deposit(300)              # Deposit into checking
    checking_account.withdraw(200)              # Withdraw from checking
    checking_account.withdraw(700)              # Attempt to withdraw more than available balance
    print(f"Checking Account Balance: ${checking_account.get_balance():.2f}")
````

### 6.Write a code for Restaurant Management System Using OOPS:

- Create a Menultem class that has attributes such as name, description, price, and category.

- Implement methods to add a new menu item, update menu item information, and remove a menu item from the menu.

- Use encapsulation to hide the menu item's unique identification number.

- Inherit from the Menultem class to create a Fooditem class and a Beverageltem class, each with their own specific attributes and methods.


````python
class MenuItem:
    def __init__(self, item_id, name, description, price, category):
        self.__item_id = item_id  # Encapsulated unique identification number
        self.name = name
        self.description = description
        self.price = price
        self.category = category

    def get_item_id(self):
        return self.__item_id

    def display_info(self):
        print(f"ID: {self.__item_id}, Name: {self.name}, Description: {self.description}, "
              f"Price: ${self.price:.2f}, Category: {self.category}")

    def update_info(self, name=None, description=None, price=None, category=None):
        if name:
            self.name = name
        if description:
            self.description = description
        if price is not None:
            self.price = price
        if category:
            self.category = category


class FoodItem(MenuItem):
    def __init__(self, item_id, name, description, price):
        super().__init__(item_id, name, description, price, category="Food")

    def display_info(self):
        print("Food Item:")
        super().display_info()


class BeverageItem(MenuItem):
    def __init__(self, item_id, name, description, price):
        super().__init__(item_id, name, description, price, category="Beverage")

    def display_info(self):
        print("Beverage Item:")
        super().display_info()


class RestaurantMenu:
    def __init__(self):
        self.menu_items = []

    def add_menu_item(self, menu_item):
        self.menu_items.append(menu_item)
        print(f"Added: {menu_item.name}")

    def update_menu_item(self, item_id, **kwargs):
        for item in self.menu_items:
            if item.get_item_id() == item_id:
                item.update_info(**kwargs)
                print(f"Updated: {item.name}")
                return
        print("Item not found.")

    def remove_menu_item(self, item_id):
        for i in range(len(self.menu_items)):
            if self.menu_items[i].get_item_id() == item_id:
                removed_item = self.menu_items.pop(i)
                print(f"Removed: {removed_item.name}")
                return
        print("Item not found.")

    def display_menu(self):
        print("\nRestaurant Menu:")
        for item in self.menu_items:
            item.display_info()


# Main program to test the Restaurant Management System
if __name__ == "__main__":
    restaurant_menu = RestaurantMenu()

    # Creating menu items
    food1 = FoodItem(1, "Burger", "Delicious beef burger", 8.99)
    food2 = FoodItem(2, "Pizza", "Cheese pizza with toppings", 12.49)
    beverage1 = BeverageItem(3, "Coke", "Refreshing cola drink", 1.99)

    # Adding menu items to the restaurant menu
    restaurant_menu.add_menu_item(food1)
    restaurant_menu.add_menu_item(food2)
    restaurant_menu.add_menu_item(beverage1)

    # Displaying the menu
    restaurant_menu.display_menu()

    # Updating a menu item
    restaurant_menu.update_menu_item(1, price=9.49)  # Update the price of the burger

    # Removing a menu item
    restaurant_menu.remove_menu_item(2)  # Remove the pizza from the menu

    # Displaying the updated menu
    restaurant_menu.display_menu()
````

````python
class Event:
    def __init__(self, event_id, name, date, time, location):
        self.__event_id = event_id  # Encapsulated unique identification number
        self.name = name
        self.date = date
        self.time = time
        self.location = location
        self.__attendees = []  # Private list of attendees

    def get_event_id(self):
        return self.__event_id

    def add_attendee(self, attendee):
        if attendee not in self.__attendees:
            self.__attendees.append(attendee)
            print(f"{attendee} has been added to the event '{self.name}'.")
        else:
            print(f"{attendee} is already an attendee of the event '{self.name}'.")

    def remove_attendee(self, attendee):
        if attendee in self.__attendees:
            self.__attendees.remove(attendee)
            print(f"{attendee} has been removed from the event '{self.name}'.")
        else:
            print(f"{attendee} is not an attendee of the event '{self.name}'.")

    def get_total_attendees(self):
        return len(self.__attendees)

    def display_event_info(self):
        print(f"Event ID: {self.__event_id}, Name: {self.name}, Date: {self.date}, "
              f"Time: {self.time}, Location: {self.location}, Total Attendees: {self.get_total_attendees()}")


class PrivateEvent(Event):
    def __init__(self, event_id, name, date, time, location, host_name):
        super().__init__(event_id, name, date, time, location)
        self.host_name = host_name

    def display_event_info(self):
        super().display_event_info()
        print(f"Host: {self.host_name}")


class PublicEvent(Event):
    def __init__(self, event_id, name, date, time, location):
        super().__init__(event_id, name, date, time, location)

    def display_event_info(self):
        super().display_event_info()
        print("This is a public event.")


class EventManagementSystem:
    def __init__(self):
        self.events = []

    def add_event(self, event):
        self.events.append(event)
        print(f"Event '{event.name}' has been created.")

    def display_events(self):
        print("\nUpcoming Events:")
        for event in self.events:
            event.display_event_info()


# Main program to test the Event Management System
if __name__ == "__main__":
    event_system = EventManagementSystem()

    # Creating events
    private_event = PrivateEvent(1, "Birthday Party", "2024-05-10", "18:00", "123 Party St.", "Alice")
    public_event = PublicEvent(2, "Community Clean-Up", "2024-06-15", "09:00", "Central Park")

    # Adding events to the system
    event_system.add_event(private_event)
    event_system.add_event(public_event)

    # Adding attendees
    private_event.add_attendee("Bob")
    private_event.add_attendee("Charlie")

    public_event.add_attendee("David")

    # Displaying all events
    event_system.display_events()

    # Removing an attendee
    private_event.remove_attendee("Bob")

    # Displaying all events after updates
    event_system.display_events()
````

### 19. Create a Python program that reads a text file named "paragraph.txt" and counts the occurrences of each word in the paragraph, displaying the results in alphabetical order.


````python
def count_word_occurrences(file_path):
    """Read the paragraph file and count the occurrences of each word."""
    word_count = {}

    try:
        with open(file_path, 'r') as file:
            # Read the entire file content
            content = file.read()
            # Split the content into words, removing punctuation and converting to lowercase
            words = content.lower().split()

            # Count occurrences of each word
            for word in words:
                # Remove punctuation from the word
                word = ''.join(char for char in word if char.isalnum())
                if word:  # Check if the word is not empty
                    if word in word_count:
                        word_count[word] += 1
                    else:
                        word_count[word] = 1

        # Sort the dictionary by keys (words) and display results
        sorted_word_count = dict(sorted(word_count.items()))
        print("Word occurrences:")
        for word, count in sorted_word_count.items():
            print(f"{word}: {count}")

    except FileNotFoundError:
        print(f"The file '{file_path}' does not exist.")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    count_word_occurrences("paragraph.txt")
````

### 20. What do you mean by Measure of Central Tendency and Measures of Dispersion. How it can be calculated.

### Measures of Central Tendency and Measures of Dispersion

**Measures of Central Tendency** are statistical metrics that describe the center or typical value of a dataset. They provide a single representative value that summarizes the entire distribution of data. The three main measures of central tendency are:

1. **Mean**: The average of all values in the dataset, calculated by summing all observations and dividing by the number of observations.
   $$
   \text{Mean} = \frac{\sum{x}}{n}
   $$
   where $$ x $$ represents each value in the dataset and $$ n $$ is the total number of values.

2. **Median**: The middle value when the data is ordered from least to greatest. If there is an even number of observations, the median is the average of the two middle numbers.

3. **Mode**: The value that appears most frequently in the dataset. A dataset may have one mode, more than one mode, or no mode at all.

### Calculation of Measures of Central Tendency

- **Mean**: Add all the numbers together and divide by how many numbers there are.
- **Median**: Sort the numbers and find the middle one (or average the two middle ones).
- **Mode**: Identify which number appears most often.

### Measures of Dispersion

**Measures of Dispersion** describe how spread out or varied the values in a dataset are. They provide insights into the variability and distribution of data points. Common measures of dispersion include:

1. **Range**: The difference between the maximum and minimum values in a dataset.
   $$
   \text{Range} = \text{Maximum} - \text{Minimum}
   $$

2. **Variance**: The average of the squared differences from the mean, indicating how much individual data points differ from the mean.
   $$
   \text{Variance} = \frac{\sum{(x_i - \mu)^2}}{n}
   $$
   where $$ x_i $$ represents each value, $$ \mu $$ is the mean, and $$ n $$ is the number of values.

### Skewness: Definition and Types

**Skewness** is a statistical measure that quantifies the asymmetry of the probability distribution of a real-valued random variable about its mean. It indicates how much and in which direction a distribution deviates from a normal distribution, which is symmetrical.

In simpler terms, skewness helps to understand whether the data points are spread out more on one side of the mean than the other. A normal distribution has a skewness of zero, meaning it is perfectly symmetrical.

### Types of Skewness

Skewness can be classified into three main types:

1. **Positive Skewness (Right Skewed)**:
   - In a positively skewed distribution, the tail on the right side (higher values) is longer or fatter than the left side.
   - This indicates that there are a few extreme high values that pull the mean to the right.
   - In this case, the relationship between mean, median, and mode is:
     $$
     \text{Mean} > \text{Median} > \text{Mode}
     $$
   - **Example**: Income distribution, where a small number of individuals earn significantly more than the majority.

   Positive Skewness

2. **Negative Skewness (Left Skewed)**:
   - In a negatively skewed distribution, the tail on the left side (lower values) is longer or fatter than the right side.
   - This indicates that there are a few extreme low values that pull the mean to the left.
   - In this case, the relationship between mean, median, and mode is:
     $$
     \text{Mean} < \text{Median} < \text{Mode}
     $$
   - **Example**: Age at retirement, where most people retire around a certain age but some retire much earlier.

   Negative Skewness

3. **Zero Skewness**:
   - A distribution with zero skewness is perfectly symmetrical. The mean, median, and mode are all equal.
   - This typically represents a normal distribution.

### Probability Mass Function (PMF) and Probability Density Function (PDF)

**Probability Mass Function (PMF)** and **Probability Density Function (PDF)** are fundamental concepts in probability theory that describe how probabilities are distributed for discrete and continuous random variables, respectively.

#### Probability Mass Function (PMF)

- **Definition**: The PMF is a function that gives the probability that a discrete random variable is exactly equal to some value. It is used to characterize the distribution of discrete random variables.
  
- **Mathematical Representation**: For a discrete random variable $$X$$, the PMF is defined as:
  $$
  P(X = x) = p_X(x)
  $$
  where $$p_X(x)$$ represents the probability that the random variable $$X$$ takes on the value $$x$$.

- **Properties**:
  - The PMF must satisfy two conditions:
    1. $$p_X(x) \geq 0$$ for all $$x$$.
    2. $$\sum_{x} p_X(x) = 1$$, meaning the total probability across all possible values must equal 1.

- **Example**: If you roll a fair six-sided die, the PMF can be represented as:
  $$
  p_X(x) = \begin{cases}
  \frac{1}{6} & \text{for } x = 1, 2, 3, 4, 5, 6 \\
  0 & \text{otherwise}
  \end{cases}
  $$

#### Probability Density Function (PDF)

- **Definition**: The PDF is a function that describes the likelihood of a continuous random variable taking on a particular value. Unlike PMF, the PDF does not give probabilities directly; instead, it gives probabilities over intervals.

- **Mathematical Representation**: For a continuous random variable $$Y$$, the PDF is defined such that:
  $$
  P(a < Y < b) = \int_a^b f_Y(y) \, dy
  $$
  where $$f_Y(y)$$ is the PDF of $$Y$$.

- **Properties**:
  - The area under the PDF curve over its entire range must equal 1:
    $$
    \int_{-\infty}^{\infty} f_Y(y) \, dy = 1
    $$
  
- **Example**: A common example is the standard normal distribution, which has a PDF given by:
  $$
  f_Y(y) = \frac{1}{\sqrt{2\pi}} e^{-\frac{y^2}{2}}
  $$

### Key Differences Between PMF and PDF

| Feature               | Probability Mass Function (PMF)          | Probability Density Function (PDF)          |
|-----------------------|-------------------------------------------|---------------------------------------------|
| Type of Variable      | Discrete                                  | Continuous                                  |
| Definition            | Gives probability for exact values        | Gives density; probabilities are found over intervals |
| Calculation           | Directly provides probabilities           | Integrates over an interval to find probabilities |
| Total Probability     | Sum of probabilities equals 1             | Area under the curve equals 1               |
| Example               | Rolling a die                             | Heights of individuals                       |

### 25. Discuss the 4 differences between correlation and regression.


Correlation and regression are both statistical methods used to analyze the relationship between variables, but they serve different purposes and have distinct characteristics. Here are four key differences between correlation and regression:

### 1. Purpose

- **Correlation**: The primary purpose of correlation is to measure the strength and direction of the relationship between two variables. It provides a single value (the correlation coefficient) that indicates how closely the two variables move together.
  
- **Regression**: The main purpose of regression is to predict the value of one variable based on the value of another. It establishes a mathematical equation that describes how changes in the independent variable (predictor) affect the dependent variable (outcome).

### 2. Relationship Type

- **Correlation**: Correlation does not imply a cause-and-effect relationship; it merely indicates that two variables are related in some way. For example, if there is a high correlation between ice cream sales and temperature, it does not mean that buying ice cream causes temperature changes.

- **Regression**: Regression analysis assumes a causal relationship, where one variable is dependent on another. For example, in a regression analysis predicting sales based on advertising spend, it is assumed that changes in advertising directly affect sales.

### 3. Interchangeability of Variables

- **Correlation**: In correlation analysis, the two variables can be interchanged without affecting the result. For instance, the correlation between variable X and variable Y is the same as that between Y and X.

- **Regression**: In regression analysis, the roles of the independent and dependent variables are fixed. Swapping these variables changes the regression model and its interpretation. For example, predicting Y from X is different from predicting X from Y.

### 4. Output

- **Correlation**: The output of correlation analysis is a single statistic known as the correlation coefficient (r), which ranges from -1 to +1. A value close to +1 indicates a strong positive correlation, while a value close to -1 indicates a strong negative correlation.

- **Regression**: The output of regression analysis includes an equation (e.g., $$Y = a + bX$$), where $$a$$ is the intercept and $$b$$ is the slope. This equation can be used for predictions, along with additional statistics such as R-squared, which indicates how well the model explains variability in the dependent variable.

**Normal Distribution**, also known as the Gaussian distribution, is a continuous probability distribution that is symmetric about the mean. It is characterized by its bell-shaped curve, which indicates that data points near the mean are more frequent than those further away. The normal distribution plays a crucial role in statistics and is widely used in various fields, including natural and social sciences, finance, and engineering.

#### Key Characteristics of Normal Distribution

- **Symmetry**: The distribution is symmetric around the mean, meaning that the left and right sides of the curve are mirror images.
- **Unimodal**: It has a single peak at the mean, which is also the median and mode.
- **Asymptotic**: The tails of the distribution approach, but never touch, the horizontal axis (x-axis), extending infinitely in both directions.
- **Empirical Rule**: Approximately 68% of data falls within one standard deviation from the mean, about 95% within two standard deviations, and about 99.7% within three standard deviations (often referred to as the 68-95-99.7 rule).

### Four Assumptions of Normal Distribution

For a dataset to be considered normally distributed, it should ideally meet the following assumptions:

1. **Independence**:
   - Observations must be independent of each other. This means that the value of one observation does not influence or provide information about another observation.
   - In practical terms, this assumption is often satisfied when data points are collected randomly from a larger population.

2. **Random Sampling**:
   - The data should be collected through random sampling methods to ensure that every member of the population has an equal chance of being included in the sample.
   - This helps in reducing bias and ensures that the sample accurately represents the population.

3. **Linearity**:
   - The relationship between variables should be linear if you are examining correlations or regressions involving normally distributed variables.
   - This means that any relationship between dependent and independent variables can be accurately represented with a straight line.

4. **Homoscedasticity**:
   - The variance among groups should be constant. In other words, the spread or dispersion of scores should be roughly equal across all levels of an independent variable.
   - This assumption is important for regression analyses and ANOVA tests where equal variances are expected for valid results.

### 31. The mean of a distribution is 60 with a standard deviation of 10. Assuming that the distribution is normal, what percentage of items be (i) between 60 and 72, (ii) between 50 and 60, (iii) beyond 72 and (iv) between 70 and 80?


````python
import scipy.stats as stats

# Given parameters
mean = 60
std_dev = 10

# Function to calculate percentage in a given range
def calculate_percentage(lower_bound, upper_bound):
    # Calculate the cumulative distribution function (CDF) values
    cdf_lower = stats.norm.cdf(lower_bound, loc=mean, scale=std_dev)
    cdf_upper = stats.norm.cdf(upper_bound, loc=mean, scale=std_dev)

    # Calculate the percentage in the range
    percentage = (cdf_upper - cdf_lower) * 100  # Convert to percentage
    return percentage

# (i) Percentage between 60 and 72
percentage_60_72 = calculate_percentage(60, 72)
print(f"Percentage of items between 60 and 72: {percentage_60_72:.2f}%")

# (ii) Percentage between 50 and 60
percentage_50_60 = calculate_percentage(50, 60)
print(f"Percentage of items between 50 and 60: {percentage_50_60:.2f}%")

# (iii) Percentage beyond 72
percentage_beyond_72 = (1 - stats.norm.cdf(72, loc=mean, scale=std_dev)) * 100
print(f"Percentage of items beyond 72: {percentage_beyond_72:.2f}%")

# (iv) Percentage between 70 and 80
percentage_70_80 = calculate_percentage(70, 80)
print(f"Percentage of items between 70 and 80: {percentage_70_80:.2f}%")
````

### 32. 15000 students sat for an examination. The mean marks was 49 and the distribution of marks had a standard deviation of 6. Assuming that the marks were normally distributed what proportion of students scored (a) more than 55 marks, (b) more than 70 marks


````python
import scipy.stats as stats

# Given parameters
mean = 49
std_dev = 6

# Function to calculate the proportion of students scoring more than a given mark
def proportion_above_score(score):
    # Calculate the Z-score
    z_score = (score - mean) / std_dev
    # Calculate the cumulative probability for the Z-score
    cumulative_prob = stats.norm.cdf(z_score)
    # Proportion of students scoring above the given score
    proportion = 1 - cumulative_prob
    return proportion

# (a) Proportion of students scoring more than 55 marks
proportion_more_than_55 = proportion_above_score(55)
print(f"Proportion of students scoring more than 55 marks: {proportion_more_than_55:.4f} ({proportion_more_than_55 * 100:.2f}%)")

# (b) Proportion of students scoring more than 70 marks
proportion_more_than_70 = proportion_above_score(70)
print(f"Proportion of students scoring more than 70 marks: {proportion_more_than_70:.4f} ({proportion_more_than_70 * 100:.2f}%)")
````

### 33. If the height of 500 students are normally distributed with mean 65 inch and standard deviation 5 inch. How many students have height: a) greater than 70 inch. b) between 60 and 70 inch.


````python
import scipy.stats as stats

# Given parameters
mean = 65       # Mean height in inches
std_dev = 5     # Standard deviation in inches
total_students = 500  # Total number of students

# Function to calculate the number of students above a certain height
def students_above_height(height):
    # Calculate the Z-score
    z_score = (height - mean) / std_dev
    # Calculate the cumulative probability for the Z-score
    cumulative_prob = stats.norm.cdf(z_score)
    # Proportion of students scoring above the given height
    proportion_above = 1 - cumulative_prob
    return proportion_above * total_students

# Function to calculate the number of students between two heights
def students_between_heights(lower_height, upper_height):
    # Calculate Z-scores for both heights
    z_lower = (lower_height - mean) / std_dev
    z_upper = (upper_height - mean) / std_dev
    # Calculate cumulative probabilities for both Z-scores
    cumulative_prob_lower = stats.norm.cdf(z_lower)
    cumulative_prob_upper = stats.norm.cdf(z_upper)
    # Proportion of students scoring between the two heights
    proportion_between = cumulative_prob_upper - cumulative_prob_lower
    return proportion_between * total_students

# (a) Number of students scoring more than 70 inches
students_more_than_70 = students_above_height(70)
print(f"Number of students scoring more than 70 inches: {students_more_than_70:.0f}")

# (b) Number of students scoring between 60 and 70 inches
students_between_60_and_70 = students_between_heights(60, 70)
print(f"Number of students scoring between 60 and 70 inches: {students_between_60_and_70:.0f}")
````

### 34. What is the statistical hypothesis? a) Explain the errors in hypothesis testing.b) Explain the Sample. What are Large Samples & Small Samples?

### Statistical Hypothesis

A **statistical hypothesis** is a specific, testable statement about a population parameter. It serves as the foundation for statistical inference and hypothesis testing. In hypothesis testing, two competing hypotheses are formulated:

1. **Null Hypothesis (H₀)**: This hypothesis states that there is no significant effect or difference, and it serves as the default assumption. For example, it might assert that the mean height of students is equal to a specific value.

2. **Alternative Hypothesis (H₁ or Hₐ)**: This hypothesis indicates the presence of an effect or difference. It represents what the researcher aims to prove. For instance, it might claim that the mean height of students is greater than a specific value.

The goal of statistical hypothesis testing is to collect data and determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

### Errors in Hypothesis Testing

In hypothesis testing, two types of errors can occur:

1. **Type I Error (α)**:
   - This error occurs when the null hypothesis is true, but we incorrectly reject it. In other words, we conclude that there is an effect or a difference when there isn't one.
   - The probability of making a Type I error is denoted by alpha (α), which is typically set at a significance level (e.g., 0.05). This means there is a 5% risk of rejecting the null hypothesis when it is actually true.

2. **Type II Error (β)**:
   - This error occurs when the null hypothesis is false, but we fail to reject it. In this case, we conclude that there is no effect or difference when there actually is one.
   - The probability of making a Type II error is denoted by beta (β). The power of a test, which is 1 - β, indicates the probability of correctly rejecting a false null hypothesis.

### Sample

A **sample** refers to a subset of individuals or observations selected from a larger population. It is used in statistical analysis to make inferences about the population without needing to collect data from every individual.

#### Types of Samples

1. **Large Samples**:
   - A sample size typically considered large is one with 30 or more observations. Large samples tend to provide more reliable estimates of population parameters due to the Central Limit Theorem, which states that the distribution of sample means approaches normality as the sample size increases.
   - Large samples reduce the impact of outliers and provide more stable estimates of population characteristics.

2. **Small Samples**:
   - A small sample usually consists of fewer than 30 observations. Small samples may not adequately represent the population, leading to less reliable estimates and greater variability.
   - When working with small samples, assumptions about the underlying population distribution become crucial, and non-parametric tests may be more appropriate if normality cannot be assumed.



### 35.A random sample of size 25 from a population gives the sample standard derivation to be 9.0. Test the hypothesis that the population standard derivation is 10.5.


````python
import scipy.stats as stats

# Given data
n = 25                      # Sample size
s = 9                       # Sample standard deviation
sigma_0 = 10.5              # Hypothesized population standard deviation
alpha = 0.05                # Significance level

# Step 1: Calculate the Chi-square statistic
s_squared = s ** 2          # Sample variance
sigma_0_squared = sigma_0 ** 2 # Hypothesized population variance

chi_square_statistic = (n - 1) * s_squared / sigma_0_squared

# Step 2: Degrees of freedom
df = n - 1

# Step 3: Critical values for two-tailed test
critical_value_lower = stats.chi2.ppf(alpha / 2, df)
critical_value_upper = stats.chi2.ppf(1 - alpha / 2, df)

# Output results
print(f"Chi-square Statistic: {chi_square_statistic:.4f}")
print(f"Critical Value Lower: {critical_value_lower:.4f}")
print(f"Critical Value Upper: {critical_value_upper:.4f}")

# Step 4: Decision
if chi_square_statistic < critical_value_lower or chi_square_statistic > critical_value_upper:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
````

### Supervised vs. Unsupervised Learning

- **Supervised Learning**:
  - Involves training a model on labeled data where both input features and output labels are provided. The model learns to map inputs to outputs based on this training data.
  
Examples: Classification (spam detection), Regression (predicting house prices).

- **Unsupervised Learning**:
  - Involves training a model on unlabeled data where only input features are provided. The model tries to identify patterns or groupings within the data without explicit labels.
  
Examples: Clustering (customer segmentation), Dimensionality Reduction (PCA).

### Bias-Variance Tradeoff

The bias-variance tradeoff is a fundamental concept in machine learning that describes the balance between two types of errors that affect model performance:

1. **Bias**:
   - The error due to overly simplistic assumptions in the learning algorithm. High bias can lead to underfitting—when a model cannot capture the underlying trend of the data.

2. **Variance**:
   - The error due to excessive complexity in the learning algorithm. High variance can lead to overfitting—when a model captures noise along with the underlying trend.

The tradeoff involves finding an optimal model complexity that minimizes total error by balancing bias and variance.

### Precision vs. Recall vs. Accuracy

1. **Precision**:
   - The ratio of true positive predictions to the total predicted positives (true positives + false positives).
   
$$
\text{Precision} = \frac{TP}{TP + FP}
$$

2. **Recall (Sensitivity)**:
   - The ratio of true positive predictions to all actual positives (true positives + false negatives).
   
$$
\text{Recall} = \frac{TP}{TP + FN}
$$

3. **Accuracy**:
   - The ratio of correctly predicted instances (true positives + true negatives) to all instances.
   
$$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$

### Overfitting and Prevention

**Overfitting** occurs when a machine learning model learns noise from the training data instead of generalizing from it. This typically happens when the model is too complex relative to the amount of training data available.

#### Prevention Techniques:
1. **Cross-validation**: Use techniques like k-fold cross-validation to ensure that your model performs well on unseen data.
2. **Regularization**: Apply techniques such as L1 or L2 regularization to penalize overly complex models.
3. **Pruning**: In decision trees, prune branches that have little importance.
4. **Early Stopping**: Stop training when performance on validation data starts to degrade.

### Cross-validation

Cross-validation is a technique used to assess how well your machine learning model generalizes to an independent dataset. It involves partitioning your dataset into subsets, training your model on some subsets while validating it on others.

The most common method is k-fold cross-validation:
1. Split the dataset into k subsets (folds).
2. For each fold:
   - Train on k-1 folds and validate on the remaining fold.
3. Average the validation scores across all folds for an overall performance metric.

### Classification vs. Regression Problems

1. **Classification Problem**:
   - A type of supervised learning problem where the output variable is categorical (discrete). The goal is to predict which category an instance belongs to based on input features.

Example: Email spam detection (spam or not spam).

2. **Regression Problem**:
   - A type of supervised learning problem where the output variable is continuous (numerical). The goal is to predict a continuous value based on input features.

Example: Predicting house prices based on various features like size and location.

### Ensemble Learning

Ensemble learning is a technique that combines multiple models (classifiers or regressors) to improve overall performance compared to individual models. The idea is that by aggregating predictions from several models, you can reduce variance (bagging), bias (boosting), or improve predictions in general.

Common ensemble methods include:
1. **Bagging (Bootstrap Aggregating)**: Reduces variance by combining predictions from multiple models trained on different subsets of data (e.g., Random Forest).
2. **Boosting**: Reduces bias by sequentially training models where each new model focuses on correcting errors made by previous ones (e.g., AdaBoost, Gradient Boosting).

### Gradient Descent

Gradient descent is an optimization algorithm used for minimizing loss functions in machine learning models by iteratively adjusting parameters in the direction of steepest descent defined by the negative gradient.

#### How it Works:
1. Initialize parameters randomly or with zeros.
2. Compute gradients of loss function concerning parameters.
3. Update parameters using:

$$
\theta = \theta - \alpha \nabla J(\theta)
$$

Where $$ \alpha $$ is the learning rate and $$ J(\theta) $$ is the loss function.

4. Repeat until convergence or for a fixed number of iterations.

### Batch Gradient Descent vs Stochastic Gradient Descent

1. **Batch Gradient Descent**:
   - Uses the entire dataset to compute gradients before updating parameters.
   - Pros: More stable convergence; better approximation of gradients.
   - Cons: Can be slow for large datasets; requires more memory.

2. **Stochastic Gradient Descent (SGD)**:
   - Updates parameters using one sample at a time randomly selected from the dataset.
   - Pros: Faster updates; can escape local minima due to noise in updates; lower memory requirements.
   - Cons: More variance in updates; may lead to oscillations around minima.

### Curse of Dimensionality

The curse of dimensionality refers to various phenomena that arise when analyzing and organizing data in high-dimensional spaces that do not occur in low-dimensional settings.

Key issues include:
1. Increased volume leading to sparse data points—making it difficult for algorithms to generalize effectively.
2. Distance metrics become less meaningful as dimensions increase—points become equidistant from each other.
3. Increased computational cost for processing high-dimensional data.

### L1 vs L2 Regularization

1. **L1 Regularization (Lasso)**:
   - Adds an absolute value penalty equal to $$ \lambda \sum |w_i| $$ where $$ w_i $$ are weights.
   - Can lead to sparse models where some weights become zero—effectively performing feature selection.

2. **L2 Regularization (Ridge)**:
   - Adds a squared penalty equal to $$ \lambda \sum w_i^2 $$.
   - Tends not to produce sparse solutions but shrinks weights uniformly—helping prevent overfitting while retaining all features.

### Confusion Matrix

A confusion matrix is a table used to evaluate the performance of a classification algorithm by comparing actual target values with those predicted by the model.

Components include:
- True Positives (TP): Correctly predicted positive cases.
- True Negatives (TN): Correctly predicted negative cases.
- False Positives (FP): Incorrectly predicted positive cases (Type I error).
- False Negatives (FN): Incorrectly predicted negative cases (Type II error).

It helps calculate various metrics like accuracy, precision, recall, and F1-score.

### AUC-ROC Curve

The AUC-ROC curve is used for evaluating binary classification models:

1. **ROC Curve**: Plots True Positive Rate (sensitivity) against False Positive Rate at various threshold settings.
2. **AUC (Area Under Curve)**: Measures how well a model distinguishes between classes; ranges from 0 to 1—where values closer to 1 indicate better performance.

### K-nearest Neighbors Algorithm

The k-nearest neighbors algorithm is a simple yet effective classification method that classifies instances based on their proximity in feature space:

1. Choose an integer $$ k $$—the number of nearest neighbors.
2. For each instance you want to classify:
    - Calculate distances between this instance and all other instances in training data using distance metrics like Euclidean distance.
    - Identify $$ k $$ closest neighbors based on calculated distances.
    - Assign class labels based on majority voting among neighbors' labels.

### Support Vector Machine Concept

A Support Vector Machine (SVM) is a supervised learning algorithm primarily used for classification tasks but can also be adapted for regression:

1. SVM aims to find an optimal hyperplane that separates classes in feature space with maximum margin between them.
2. It uses support vectors—data points closest to decision boundary—to define this hyperplane effectively while minimizing classification error through techniques like kernel trick for non-linear separations.


### How the Kernel Trick Works in SVM

The kernel trick is a technique used in Support Vector Machines (SVM) to enable the classification of non-linear data by transforming it into a higher-dimensional space where it becomes linearly separable. Instead of explicitly mapping the input features into this higher-dimensional space, the kernel trick allows us to compute the dot products of the transformed data directly using a kernel function.

**How It Works**:
1. **Feature Mapping**: The kernel trick implicitly maps input data points into a higher-dimensional feature space using a kernel function $$ K(x, y) $$, which computes the inner product of two points in this new space.
2. **Linear Separation**: In this higher-dimensional space, a linear hyperplane can be used to separate classes that are not linearly separable in the original feature space.
3. **Efficiency**: The kernel trick avoids the computational cost of transforming data explicitly, making it feasible to work with high-dimensional spaces.

### Different Types of Kernels Used in SVM

1. **Linear Kernel**:
   - **Formula**: $$ K(x, y) = x^T y $$
   - **Use Case**: Suitable for linearly separable data.

2. **Polynomial Kernel**:
   - **Formula**: $$ K(x, y) = (x^T y + c)^d $$ where $$ c $$ is a constant and $$ d $$ is the degree of the polynomial.
   - **Use Case**: Useful for datasets where the relationship between features is polynomial.

3. **Radial Basis Function (RBF) Kernel (Gaussian Kernel)**:
   - **Formula**: $$ K(x, y) = \exp\left(-\frac{\|x - y\|^2}{2\sigma^2}\right) $$
   - **Use Case**: Effective for non-linear data; it can handle cases where the decision boundary is very complex.

4. **Sigmoid Kernel**:
   - **Formula**: $$ K(x, y) = \tanh(\alpha x^T y + c) $$
   - **Use Case**: Mimics neural networks; less commonly used than others.

### Hyperplane in SVM

The hyperplane is a decision boundary that separates different classes in SVM. It is determined by maximizing the margin between the closest points of different classes (support vectors). The equation of a hyperplane in an n-dimensional space can be expressed as:

$$ w^T x + b = 0 $$

Where:
- $$ w $$ is the weight vector (normal to the hyperplane).
- $$ x $$ represents input features.
- $$ b $$ is the bias term.

### Pros and Cons of Using SVM

#### Pros:
- Effective in high-dimensional spaces.
- Works well with clear margin separation.
- Memory efficient since it uses a subset of training points (support vectors).
- Versatile with different kernel functions.

#### Cons:
- Not suitable for large datasets due to high training time.
- Less effective on noisy datasets with overlapping classes.
- Choosing an appropriate kernel and tuning parameters can be challenging.

### Hard Margin vs Soft Margin SVM

1. **Hard Margin SVM**:
   - Assumes that data is perfectly linearly separable.
   - No misclassifications are allowed; all data points must lie outside the margin.
   - Sensitive to outliers; may not work well with real-world noisy data.

2. **Soft Margin SVM**:
   - Allows some misclassifications; introduces slack variables to handle them.
   - Balances maximizing margin and minimizing classification error.
   - More robust to noise and outliers, making it suitable for real-world applications.

### Constructing a Decision Tree

1. **Select Best Feature**: Choose the feature that best splits the data based on a certain criterion (e.g., Gini impurity, information gain).
2. **Split Data**: Divide the dataset into subsets based on the chosen feature's values.
3. **Repeat Process**: Recursively apply steps 1 and 2 on each subset until stopping criteria are met (e.g., maximum depth reached, no further splits improve purity).
4. **Assign Labels**: Assign labels to leaf nodes based on majority class or mean value for regression tasks.

### Working Principle of a Decision Tree

A decision tree works by recursively splitting the dataset into subsets based on feature values until reaching a stopping condition (leaf nodes). Each internal node represents a feature test, each branch corresponds to an outcome of that test, and each leaf node represents a class label or value prediction.

### Information Gain in Decision Trees

Information gain measures how much information a feature provides about the class label. It quantifies the reduction in entropy or uncertainty when splitting on that feature:

$$
\text{Information Gain} = H(S) - H(S|A)
$$

Where:
- $$ H(S) $$: Entropy of the original dataset.
- $$ H(S|A) $$: Weighted entropy after splitting on attribute $$ A $$.

Used to select features that provide maximum information about class labels during tree construction.

### Gini Impurity in Decision Trees

Gini impurity measures how often a randomly chosen element from the set would be incorrectly labeled if it was randomly labeled according to the distribution of labels in the subset:

$$
Gini(S) = 1 - \sum_{i=1}^{C} p_i^2
$$

Where $$ p_i $$ is the proportion of instances belonging to class $$ i $$. Lower Gini impurity indicates better purity.

### Advantages and Disadvantages of Decision Trees

#### Advantages:
- Easy to interpret and visualize.
- Requires little data preprocessing (no need for normalization).
- Can handle both numerical and categorical data.
- Non-parametric; does not assume any underlying distribution.

#### Disadvantages:
- Prone to overfitting, especially with deep trees.
- Sensitive to small variations in data; slight changes can lead to different trees.
- Can create biased trees if some classes dominate.

### How Random Forests Improve Upon Decision Trees

Random forests improve upon decision trees by combining multiple decision trees (ensemble learning), which helps reduce overfitting and increases accuracy:

1. **Bootstrap Aggregating (Bagging)**: Each tree is trained on a random subset of data drawn with replacement from the training set.
2. **Feature Randomness**: At each split, only a random subset of features is considered, promoting diversity among trees and reducing correlation.

### How Random Forest Algorithm Works

1. Create multiple decision trees using bootstrapped samples from training data.
2. For each tree, select random subsets of features at each split point.
3. Aggregate predictions from all trees (majority voting for classification or averaging for regression).

### Bootstrapping in Random Forests

Bootstrapping involves sampling with replacement from the training dataset to create multiple subsets for training individual trees in the random forest. This technique helps ensure diversity among trees while maintaining a robust model performance.

### Concept of Feature Importance in Random Forests

Feature importance measures how much each feature contributes to improving model accuracy across all trees in the forest. It helps identify which features are most influential for predictions and can guide feature selection or engineering efforts.

### Key Hyperparameters of Random Forests

1. **n_estimators**: Number of trees in the forest; more trees generally improve performance but increase computation time.
2. **max_depth**: Maximum depth of each tree; controls overfitting—deeper trees capture more detail but may lead to overfitting.
3. **min_samples_split**: Minimum number of samples required to split an internal node; higher values prevent creating nodes with few samples.
4. **max_features**: Number of features considered at each split; influences diversity among trees—can be set as an integer, float, or string ("auto", "sqrt", "log2").

### Logistic Regression Model and Its Assumptions

Logistic regression is a statistical model used for binary classification problems that predicts probabilities using a logistic function:

1. **Logit Function**:
   - Models log odds as a linear combination of input features:

$$
\log\left(\frac{p}{1-p}\right) = \beta_0 + \beta_1 x_1 + ... + \beta_n x_n
$$

Where $$ p $$ is the probability of class 1.

#### Assumptions:
1. The dependent variable is binary.
2. Observations are independent.
3. There’s little or no multicollinearity among predictors.
4. The relationship between independent variables and log odds is linear.

### Handling Binary Classification Problems with Logistic Regression

Logistic regression handles binary classification by modeling probabilities using logistic functions, allowing predictions between 0 and 1:

1. Fit model using training data.
2. Predict probabilities for test instances using learned coefficients.
3. Classify instances based on thresholding predicted probabilities (commonly at 0.5).

### Sigmoid Function in Logistic Regression

The sigmoid function transforms any real-valued number into a value between 0 and 1:

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

In logistic regression, it converts log odds from linear combinations into probabilities suitable for binary classification.

### Cost Function in Logistic Regression

The cost function measures how well logistic regression predicts outcomes compared to actual values:

$$
J(\beta) = -\frac{1}{m} \sum_{i=1}^{m} [y^{(i)} \log(h_\beta(x^{(i)})) + (1-y^{(i)})\log(1-h_\beta(x^{(i)}))]
$$

Where:
- $$ m $$: Number of observations
- $$ h_\beta(x^{(i)}) $$: Predicted probability for instance $$ i $$

The goal is to minimize this cost function through optimization techniques like gradient descent.

### Extending Logistic Regression for Multiclass Classification

Logistic regression can be extended using strategies like:

1. **One-vs-Rest (OvR)**: Train one classifier per class; predict class probabilities and choose class with highest probability as final prediction.
  
2. **Softmax Regression**: Generalization of logistic regression for multiclass problems using softmax function instead of sigmoid:

$$
P(y=k|x) = \frac{e^{\beta_k^T x}}{\sum_{j=1}^{K} e^{\beta_j^T x}}
$$

Where $$ K $$ is number of classes.

### L1 vs L2 Regularization in Logistic Regression

1. **L1 Regularization (Lasso)**:
   - Adds absolute value penalty equal to $$ \lambda ||w||_1 $$ where $$ w $$ are weights.
   - Encourages sparsity—some weights become zero—effectively performing feature selection.

2. **L2 Regularization (Ridge)**:
   - Adds squared penalty equal to $$ \lambda ||w||_2^2 $$ where $$ w $$ are weights.
   - Shrinks weights uniformly but retains all features—helps prevent overfitting without eliminating any features entirely.

### XGBoost Overview

XGBoost (Extreme Gradient Boosting) is an optimized implementation of gradient boosting framework designed for speed and performance improvements over traditional boosting algorithms.

#### Key Differences from Other Boosting Algorithms:
- Implements regularization techniques (L1 & L2).
- Uses parallel processing during tree construction for faster computation.
- Includes built-in handling for missing values during training without requiring imputation beforehand.

### Concept of Boosting in Ensemble Learning

Boosting refers to an ensemble learning technique that combines multiple weak learners (typically decision trees) into a single strong learner by sequentially training models where each new model focuses on correcting errors made by previous ones:

1. Each model is trained on weighted versions of instances—more weight given to previously misclassified instances.
2. Predictions from all models are combined through weighted voting or averaging.

### Handling Missing Values in XGBoost

XGBoost has built-in mechanisms for handling missing values during training:

- It automatically learns what happens when it encounters missing values at each split point and assigns them to either side based on minimizing loss functions—effectively treating them as separate branches without needing explicit imputation beforehand.

### Key Hyperparameters in XGBoost

Some important hyperparameters include:

1. **n_estimators**: Number of boosting rounds or trees; more can lead to better performance but increase computation time.
  
2. **learning_rate (eta)**: Step size shrinkage used in update steps; smaller values require more boosting rounds but lead to better generalization.

3. **max_depth**: Maximum depth of trees; deeper trees capture more complexity but may lead to overfitting if too deep without sufficient regularization.

4. **subsample**: Fraction of samples used per tree; lower values help prevent overfitting by introducing randomness but may reduce performance if too low.

5. **colsample_bytree/colsample_bylevel/colsample_bynode**: Control subsampling ratio for columns at various stages—helps reduce overfitting by introducing randomness among features considered at splits.

### Gradient Boosting Process in XGBoost

Gradient boosting involves iteratively adding weak learners (trees):

1. Initialize predictions with mean value or constant prediction based on target variable distribution.
  
2. For each iteration:
   - Compute residual errors from previous predictions as target variable minus predicted value.
   - Fit new weak learner (tree) on these residuals as target variable using gradient descent principles—minimizing loss function concerning predictions made so far.
  
3. Update overall predictions by adding weighted contributions from new learner until stopping criteria are met (e.g., maximum iterations reached).
