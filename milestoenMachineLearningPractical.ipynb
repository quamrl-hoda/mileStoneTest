```python
# Filter data for patients with lung cancer
lung_cancer_data = df[df['LUNG_CANCER'] == 'YES']

# Count occurrences of each symptom
symptom_columns = ['FATIGUE ', 'ALLERGY ', 'WHEEZING', 'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']
symptom_counts = {symptom: lung_cancer_data[symptom].value_counts() for symptom in symptom_columns}

# Display counts for each symptom
for symptom, counts in symptom_counts.items():
    print(f"\nCounts for {symptom}:")
    print(counts)
```

```python
# Set up a figure for plotting
plt.figure(figsize=(14, 8))

# Plot each symptom's counts
for i, symptom in enumerate(symptom_columns):
    plt.subplot(3, 3, i + 1)
    sns.countplot(data=lung_cancer_data, x=symptom, order=lung_cancer_data[symptom].value_counts().index)
    plt.title(f'Distribution of {symptom}')
    plt.xlabel(symptom)
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Display unique values for 'CHEST PAIN' and 'LUNG_CANCER'
print("Unique values in CHEST PAIN:", df['CHEST PAIN'].unique())
print("Unique values in LUNG CANCER:", df['LUNG_CANCER'].unique())
```

```python
# Create a contingency table
contingency_table = pd.crosstab(df['CHEST PAIN'], df['LUNG_CANCER'])
print("Contingency Table:")
print(contingency_table)

# Calculate percentages
contingency_percentage = contingency_table.div(contingency_table.sum(axis=1), axis=0) * 100
print("\nPercentage Table:")
print(contingency_percentage)
```

```python
# Plotting the correlation between chest pain and lung cancer diagnosis
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='CHEST PAIN', hue='LUNG_CANCER')
plt.title('Correlation between Chest Pain and Lung Cancer Diagnosis')
plt.xlabel('Chest Pain')
plt.ylabel('Count')
plt.legend(title='Lung Cancer', loc='upper right', labels=['No', 'Yes'])
plt.show()
```

```python
# Display unique values for relevant symptom columns
symptom_columns = ['FATIGUE ', 'ALLERGY ', 'WHEEZING', 'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN']
for symptom in symptom_columns:
    print(f"Unique values in {symptom}:", df[symptom].unique())
```

```python
# Create a function to generate contingency tables for symptoms vs lung cancer
def create_contingency_table(symptom):
    return pd.crosstab(df[symptom], df['LUNG_CANCER'], margins=True)

# Analyze each symptom in relation to lung cancer
contingency_tables = {}
for symptom in symptom_columns:
    contingency_tables[symptom] = create_contingency_table(symptom)
    print(f"\nContingency Table for {symptom}:")
    print(contingency_tables[symptom])
```

```python
# Calculate percentages for each symptom's contingency table
percentage_tables = {}
for symptom, table in contingency_tables.items():
    percentage_tables[symptom] = table.div(table.sum(axis=1), axis=0) * 100
    print(f"\nPercentage Table for {symptom}:")
    print(percentage_tables[symptom])
```

```python
# Set up a figure for plotting
plt.figure(figsize=(14, 10))

# Plotting each symptom's correlation with lung cancer diagnosis
for i, symptom in enumerate(symptom_columns):
    plt.subplot(3, 3, i + 1)
    sns.countplot(data=df, x=symptom, hue='LUNG_CANCER')
    plt.title(f'Correlation between {symptom} and Lung Cancer Diagnosis')
    plt.xlabel(symptom)
    plt.ylabel('Count')
    plt.legend(title='Lung Cancer', loc='upper right', labels=['No', 'Yes'])

plt.tight_layout()
plt.show()
```

### age range that makes up the majority of voters

```python
# Define age bins and labels
bins = [18, 25, 35, 45, 55, 65, 75, 85, 95]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65-74', '75-84', '85+']

# Create a new column for age groups
data['age_group'] = pd.cut(data['age'], bins=bins, labels=labels, right=False)

# Count occurrences of each age group
age_group_counts = data['age_group'].value_counts().sort_index()
print("\nAge Group Counts:")
print(age_group_counts)

# Find the most common age group
most_common_age_group = age_group_counts.idxmax()
most_common_count = age_group_counts.max()

print(f"\nThe most common age range among voters is: {most_common_age_group} with {most_common_count} voters.")
```
